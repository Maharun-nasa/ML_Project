## 1. Regression 
> Here we want to predict the `Gross Income` based on the `Price` and `Tax` of the products . For this we use the multiple regression model .

```python
# Import necessary libraries 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score , mean_absolute_error
```

## Pre-processing the data before applying ML model .

```python
df1 = df[['Unit price','Tax 5%','gross income']] # Assigning the selected columns to a new dataframe
df1
```

```python
df1[['Unit price','Tax 5%','gross income']].isnull().sum() # Checking for missing values in the selected columns 
```

```python
# checking the distributiopns of the data 
plt.figure(figsize=(24,6))
plt.subplot(1,4,1)
sns.histplot(df1['Unit price'],kde=True)
plt.subplot(1,4,2)
sns.histplot(df1['Tax 5%'],kde = True)
```

### Assumption of Regression is : independent variable must be normally distributed . So, Now we normalize the data by using Quantile transformation

```python
# normalizing the unit price & Tax column using quantile transformation
from sklearn.preprocessing import QuantileTransformer

qt = QuantileTransformer(output_distribution='normal')
df1['Unit price'] = qt.fit_transform(df1[['Unit price']])
df1['Tax 5%'] = qt.fit_transform(df1[['Tax 5%']])

# plot the distributions after normalization 
plt.figure(figsize = (24,6))
plt.subplot(1,4,1)
sns.histplot(df1['Unit price'],kde=True)
plt.subplot(1,4,2)
sns.histplot(df1['Tax 5%'],kde = True)
plt.show() 
```

```python
# check correlation between price and tax
sns.scatterplot(x='Unit price', y= "Tax 5%", data=df1)
#correlation value
cor = df1['Unit price'].corr(df1['Tax 5%']) ; cor
```

```python
# Checking for outliers in the selected columns using boxplots 
plt.figure(figsize=(20,8))
plt.subplot(1,4,1)
sns.boxplot(df1['Unit price'])
plt.subplot(1,4,2)
sns.boxplot(df1['Tax 5%'])
plt.subplot(1,4,3)
sns.boxplot(df1['gross income'])
plt.show()
```

```python
# for multiple linear regression define the independent variables (x) and dependent variable (y)
x = df1[['Unit price','Tax 5%']]
y = df1['gross income']

x_train,x_test,y_train,y_test=train_test_split(x,y,train_size = 0.8, random_state = 42)

#model
model=LinearRegression()

#train the model
model.fit(x_train,y_train)

#Predict the train model
predictions= model.predict(x_test)

#compare the actual vs predicted values
sns.regplot(x=y_test, y=predictions, color = 'darkgreen' , scatter_kws={'color': 'orange'})

# Evaluate the model
mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test,predictions)
mae = mean_absolute_error(y_test,predictions)
print ("MSE:",mse) ; print ("R2 Score:",r2) ; print("MAE:",mae)
```

### **Interpretation : Linear Regression Model Training & Evaluation**

1. **Train-Test Split**: Divides data into 80% training and 20% testing
2. **Model Creation**: Initializes LinearRegression object from sklearn
3. **Model Training**: Fits the model on training data
4. **Predictions**: Makes predictions on test data
5. **Visualization**: Creates scatter plot comparing actual vs predicted values

- **MSE**: Average of squared differences between predicted and actual values.Here MSE = 20.128 indicates that the model's predictions are relatively close to the actual values . 
- **R2 Score** : Proportion of variance in target explained by the model . Here r2 = 0.86 suggests that the model explains 86% of the variance in the dependent variable, which is a good fit. 
- **MAE** : Absolute difference between predicted and actual values . Here MAE = 3.047 indicates that on average tye model's predictions are off by 3.047 units from the actual values .
> Overall, these metrics suggest that the model is performing well in predicting gross income based on unit price and tax. 

## 2. Logistic Regression 
### Here we want to **predict customer satisfaction level** by their provided `Rating` based on the  `Unit_price` , `Payment`, `Quantity` , `Gender` , `Product_line` , `City` & `Customer Type` .

> For this we have to convert the Rating variable into binary outcomes (as  High rating = 1 if >=7, Low rating = 0 if <7) because logistic regression uses binary dependent variable for predicting . 

```python
# convert the continuous variable Rating into binary outcomes 
df['High rating'] = (df['Rating']>=7).astype(int) #astype(int) is use to convert the boolean values into 0 and 1 
```

```python
# assigning the selected columns to a new dataframe so that the original dataframe will not be affected by the changes we make in the new dataframe.
df2 = df[['Unit price','Gender','Product line','City','Customer type','Quantity','Payment','High rating']]
df2.head()
```

```python
df2['Unit price'] = df1['Unit price']
sns.histplot(x=df2['Unit price'], kde = True)
```

```python
# Encoding the Categorical variables for numerical analysis . 

from sklearn.preprocessing import LabelEncoder 
le=LabelEncoder()

df2['Gender']=le.fit_transform(df2['Gender'])
df2['City']=le.fit_transform(df2['City'])
df2['Product line']=le.fit_transform(df2['Product line']) 
df2['Payment']= le.fit_transform(df2['Payment'])
df2['Customer type']= le.fit_transform(df2['Customer type'])
df2
```

```python
# define the feature and target variable for classification model. 
x = df2[[ 'Unit price','Gender','Product line','City','Customer type','Quantity','Payment']]
y = df2['High rating']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, accuracy_score,recall_score ,confusion_matrix

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2 ,random_state=42)

model=LogisticRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)

cm=confusion_matrix(y_test,y_pred)
print("confusion matrix : ",cm)

sns.heatmap(cm,annot=True)
plt.title('CM of Logistic Regression')
plt.ylabel('predicted label')
plt.xlabel('Actual label')
plt.show()

print("Accuracy Score :",accuracy_score(y_test,y_pred))
print("precision_score: ",precision_score(y_test,y_pred))
print("recall_score: ",recall_score(y_test,y_pred))
```

### **Interpretation :** Logistic Regression Model Shows Very Poor performance to predict Customer Satisfaction Level . Precision Score , recall score , accuracy score Shows very low <50% .
