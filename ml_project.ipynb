## 1. Regression
> Here we want to predict the `Gross Income` based on the `Price` and `Tax` of the products . For this we use the multiple regression model .

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df1 = df[['Unit price','Tax 5%','gross income']]
df1
```

```python
sns.histplot(df1['gross income'],kde = True)
```

```python
# for multiple linear regression define the independent variables (x) and dependent variable (y)
x = df1[['Unit price','Tax 5%']]
y = df1['gross income']

x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=.8)

#model
model=LinearRegression()

#train the model
model.fit(x_train,y_train)

#Predict the train model
predictions= model.predict(x_test)

#compare
sns.scatterplot(x=y_test,y=predictions)

# Evaluate the model
mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test,predictions)
print ("MSE:",mse)
print ("R2 Score:",r2)
```

### **Interpretation : Linear Regression Model Training & Evaluation**

1. **Train-Test Split**: Divides data into 80% training and 20% testing
2. **Model Creation**: Initializes LinearRegression object from sklearn
3. **Model Training**: Fits the model on training data
4. **Predictions**: Makes predictions on test data
5. **Visualization**: Creates scatter plot comparing actual vs predicted values

- **MSE**: Average of squared differences between predicted and actual values.Here MSE Close to 0 Perfect predictions, model errors are negligible .
- **R2 Score** : Proportion of variance in target explained by the model . Here r2 = 1 indicates perfect predictions .This is not indicative of real-world prediction problems because model's performance is artificially inflated due to mathematical relationship.

```python
df['High rating'].value_counts()
```

```python
# assigning the selected columns to a new dataframe so that the original dataframe will not be affected by the changes we make in the new dataframe.
df2 = df[['Unit price','Gender','Product line','City','Customer type','Quantity','Payment','High rating']]
df2.head()
```

```python
# Encoding the Categorical variables for numerical analysis .

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

df2['Gender']=le.fit_transform(df2['Gender'])
df2['City']=le.fit_transform(df2['City'])
df2['Product line']=le.fit_transform(df2['Product line'])
df2['Payment']= le.fit_transform(df2['Payment'])
df2['Customer type']= le.fit_transform(df2['Customer type'])
df2
```

```python
x = df2[[ 'Unit price','Gender','Product line','City','Customer type','Quantity','Payment']]
y = df2['High rating']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, accuracy_score,recall_score ,confusion_matrix

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2 ,random_state=42)

model=LogisticRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)

cm=confusion_matrix(y_test,y_pred)
print("confusion matrix : ",cm)

sns.heatmap(cm,annot=True)
plt.ylabel('Predicted label')
plt.xlabel('Actual label')
plt.show()

print("Accuracy Score :",accuracy_score(y_test,y_pred))
print("precision_score: ",precision_score(y_test,y_pred))
print("recall_score: ",recall_score(y_test,y_pred))
```

## 5. KNN (K-Nearest Neighbors)
> Predict customer satisfaction level (High/Low Rating) using KNN algorithm.
Using the same `x` and `y` variables as Logistic Regression, Decision Tree & Random Forest.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
import joblib

# Re-define x, y using the same variables as Logistic Regression, Decision Tree & Random Forest
x = df2[['Unit price','Gender','Product line','City','Customer type','Quantity','Payment']]
y = df2['High rating']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Feature Scaling (important for KNN - distance based algorithm)
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

print('Feature scaling completed successfully.')
print(f'x_train shape: {x_train_scaled.shape}')
print(f'x_test shape: {x_test_scaled.shape}')
print(f'y_train unique values: {sorted(y_train.unique())}')
```

```python
# Train KNN Classifier with k=5
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train_scaled, y_train)

# Predict
y_pred_knn = knn.predict(x_test_scaled)

# Evaluate
cm_knn = confusion_matrix(y_test, y_pred_knn)
print('KNN Confusion Matrix (k=5):')
print(cm_knn)

plt.figure(figsize=(6, 4))
sns.heatmap(cm_knn, annot=True, fmt='d', cmap='Blues')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.title('KNN Confusion Matrix (k=5)')
plt.show()

print(f'\nKNN Accuracy Score: {accuracy_score(y_test, y_pred_knn):.4f}')
print(f'KNN Precision Score: {precision_score(y_test, y_pred_knn):.4f}')
print(f'KNN Recall Score: {recall_score(y_test, y_pred_knn):.4f}')
print('\nClassification Report:')
print(classification_report(y_test, y_pred_knn))
```

**Reasoning**:
To find the optimal K, we test K=1 to K=20 and pick the one with the lowest error rate.


```python
# Finding the best K value
error_rate = []
k_range = range(1, 21)

for k in k_range:
    knn_temp = KNeighborsClassifier(n_neighbors=k)
    knn_temp.fit(x_train_scaled, y_train)
    y_pred_temp = knn_temp.predict(x_test_scaled)
    error_rate.append(np.mean(y_pred_temp != y_test))

plt.figure(figsize=(10, 6))
plt.plot(list(k_range), error_rate, marker='o', linestyle='--', color='red')
plt.title('Error Rate vs K Value')
plt.xlabel('K Value')
plt.ylabel('Error Rate')
plt.xticks(list(k_range))
plt.grid(True)
plt.show()

best_k = list(k_range)[error_rate.index(min(error_rate))]
print(f'Best K value: {best_k} with Error Rate: {min(error_rate):.4f}')
```

```python
# Retrain with best K
knn_best = KNeighborsClassifier(n_neighbors=best_k)
knn_best.fit(x_train_scaled, y_train)

y_pred_best = knn_best.predict(x_test_scaled)
print(f'Best KNN Model (K={best_k})')
print(f'Accuracy: {accuracy_score(y_test, y_pred_best):.4f}')
print(f'Precision: {precision_score(y_test, y_pred_best):.4f}')
print(f'Recall: {recall_score(y_test, y_pred_best):.4f}')
print('\nClassification Report:\n', classification_report(y_test, y_pred_best))
```

```python
# Save the KNN model and scaler
joblib.dump(knn_best, 'knn_model.pkl')
joblib.dump(scaler, 'knn_scaler.pkl')

print("Model saved as 'knn_model.pkl'")
print("Scaler saved as 'knn_scaler.pkl'")
```

The **KNN (K-Nearest Neighbors)** model predicts customer satisfaction (High/Low Rating) using the same features as other models.

- **Feature Scaling** applied using `StandardScaler` because KNN is distance-based.
- **Best K** selected by testing K=1 to 20, choosing the lowest error rate.
- Model saved as `knn_model.pkl` and scaler saved as `knn_scaler.pkl` using `joblib`.
